```python
# Import dependencies
import pandas as pd
import os
```

```python
# Read file into DataFrame
#team_one = os.path.join('../','Resources', 'pools.csv')
#df= pd.read_csv(team_one)
pool_df = pd.read_csv('../Resources/ratings_and_sentiments.csv')
pool_df
```

```python
# Find all reviews with a number rating greater than or equal to 4.
#rating= (pool_df['num_rating']>=4).value_counts()
rating_df= pool_df[pool_df['num_rating']>=4]
len(rating_df)
```

```python
# Find all the coffee shops with a overall sentiment greater than or equal to 4.
overall_df = pool_df[pool_df['overall_sent']>=4]
len(overall_df)
```

```python
len(pool_df[(pool_df['overall_sent']>=4) & (pool_df['num_rating']>=4)])
```

```python
# Return the count of the above query.
len(pool_df[(pool_df['overall_sent'] >= 4) & (pool_df['num_rating'] >= 4)])
```

```python
# Change the name of the review_text column to review
pool_df.rename (columns = {'review_text':'review'})
```

```python
# Change the name of the bool_HIGH column to bool_high
pool_df.rename(columns = {'bool_HIGH':'bool_high'})
```

```python
# Save your changes to the file as a pipe delimited file.
pool_df.to_csv('pool_df',sep= '|')
```
